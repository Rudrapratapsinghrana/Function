Q1->  What is the difference between a function and a method in Python?

  =>A function is a standalone block of code, while a method is a function
   associated with an object and invoked on it.

Q2-> Explain the concept of function arguments and parameters in Python?

  ->Parameters are placeholders defined in a function, while arguments are   actual values passed to the function during a call.

Q3->What are the different ways to define and call a function in Python?

 ->Functions in Python can be defined using 'def' or 'lambda' and called by their name with appropriate arguments in parentheses.

Q4-> What is the purpose of the `return` statement in a Python function?

 ->This function sends a result or value back to the caller, ending the function's execution.

Q5-> What are iterators in Python and how do they differ from iterables?

 ->An iterable is an object that can be looped over, while an iterator is an object created from an iterable that produces items one at a time using '__next__()'.

Q6-> Explain the concept of generators in Python and how they are defined?

  ->they are special iterators defined using a function with 'yield' to produce values lazily, one at a time.

Q7-> What are the advantages of using generators over regular functions?

 ->Generators save memory and improve performance by yielding values lazily, generating them on demand instead of storing them in memory.

Q8-> What is a lambda function in Python and when is it typically used?

 ->it is a single-expression function typically used for short, simple operations without defining a full function.

Q9->  Explain the purpose and usage of the `map()` function in Python.

  ->It applies a given function to each item in an iterable, returning a map object with the results.

Q10->. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 ->map() applies a function to all items in an iterable, filter() selects items that meet a condition, and reduce() combines items into a single value using a function.
